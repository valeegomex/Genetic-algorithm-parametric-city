```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
plt.style.use('bmh')
plt.rcParams.update({'font.size': 16})
```

```python
# Graficar rendimientos por separado
def graficar_min_mean(y_min, y_mean, title, ax):
       x = np.arange(0, len(y_min), 1)
       # Data for plotting
       ax.plot(x, y_min)
       ax.scatter(x, y_min, label='MVRC minimal población')
       ax.plot(x, y_mean)
       ax.scatter(x, y_mean, label='MVRC promedio población')
       ax.legend()
       ax.set(xlabel='Generación', ylabel='MVRC',
              title=title)
       ax.xaxis.set_major_locator(MaxNLocator(integer=True))
       return
```

```python
def agregar_comparacion(y_min, y_mean, label, ax1, ax2, color):
       x =  np.arange(0, len(y_min), 1)
       # ax 1 es el minimo
       ax1.plot(x, y_min, color=color)
       ax1.scatter(x, y_min, label=label, color=color)
       ax2.plot(x, y_mean, color=color)
       ax2.scatter(x, y_mean, label=label, color=color)
       return
```

## Numero de líneas graficos

```python
import math


def cantidad_lineas_con_periferia(n_zonas:int):

    total = 0
    numerador = math.factorial(2*n_zonas + 1)
    for l in range(2, 2*n_zonas + 1):
        denominador = math.factorial(2*n_zonas + 1 - l)
        if denominador > 0:
            total += numerador/denominador

    return total

cantidad_lineas = []
for n_zonas in range(1, 11):
    cantidad_lineas.append(cantidad_lineas_con_periferia(n_zonas))

cantidad_lineas
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14})

# Datos
x = [i for i in range(1, 11)]
y = cantidad_lineas

# Crear gráfico
plt.scatter(x, y, c='purple')

# Configuración de ejes en escala logarítmica
plt.yscale('log')

# Etiquetas y título
plt.xlabel('Parametric City zones')
plt.ylabel('Number of lines')
plt.xlim(-0.5, 10.5)
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.savefig('Imagenes/cantidad_lineas_con_periferia.png', dpi=400)
plt.show()
```
